# Описание проекта
Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Вам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.
Перед вами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и вы планируете кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируете ли вы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.
В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».


### Описание данных
1. Name — название игры
2. Platform — платформа
3. Year_of_Release — год выпуска
4. Genre — жанр игры
5. NA_sales — продажи в Северной Америке (миллионы проданных копий)
6. EU_sales — продажи в Европе (миллионы проданных копий)
7. JP_sales — продажи в Японии (миллионы проданных копий)
8. Other_sales — продажи в других странах (миллионы проданных копий)
9. Critic_Score — оценка критиков (максимум 100)
10. User_Score — оценка пользователей (максимум 10)
11. Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

```python
import pandas as pd
print(f'Версия pandas: {pd.__version__}')
import numpy as np
print(f'Версия numpy: {np.__version__}')
import seaborn as sns
print(f'Версия seaborn: {sns.__version__}')
from scipy import stats as st
import matplotlib as mpl
from matplotlib import pyplot as plt
from matplotlib.patches import ConnectionPatch
```

```python
try:
    data = pd.read_csv('games.csv')
except:
    data = pd.read_csv('/datasets/games.csv')    
```

### Подготовка данных

```python
data.head()
```

```python
data.info()
```

```python
# columns to lowercase
data.columns = data.columns.str.lower()
```

Избавимся от строк с пропущенными значениями в столбцах 'name' и 'year_of_release'. Количество строк очень мало, поэтому это не повлияет на итоговые результаты.

```python
data.dropna(subset=['name', 'year_of_release'], inplace=True)
```

```python
# year_of_release to int
data.year_of_release = data.year_of_release.astype('int64')
```

Проверим на дубликаты genre и rating

```python
data.genre.value_counts()
```

```python
data.rating.value_counts()
```

На [вики](https://ru.wikipedia.org/wiki/Entertainment_Software_Rating_Board) сказано, что K A - это устаревшее обозначение E. Заменим K-A на E.

```python
data.loc[data['rating'] == 'K-A', 'rating'] = 'E'
```

```python
data[data.rating.isna()==True]
```

Пропуски в колонках critic_score, user_score и rating могли появится в следствие невысокой популярности игр. Для некоторых игр это можно наблюдать и по продажам.
Однако, эти строки удалять не станем, т.к. в них есть информация, которая может повлиять на результат исследования.

```python
data.user_score.value_counts()
```

tbd может расшифровываться по-разному. Чаще всего это to be determined - еще не определен. Инфо [от сюда](https://autogear.ru/article/171/345/tbd---chto-znachit-eto-sokraschenie/).

```python
data.query('user_score == "tbd"').pivot_table(index='platform',
                                             values='user_score',
                                             aggfunc='count').sort_values(by='user_score', ascending=False).head()
```

По факту, значения tbd - это неопределенные, пропущенные значения. Однако, в этих строках так же содержатся много полезной информации, которая может пригодиться в дальнейшем исследовании. Заменим tbd на NaN

```python
data.loc[data['user_score'] == 'tbd', 'user_score'] = np.nan
```

```python
data.user_score = data.user_score.astype('float64')
```

Посчитаем суммарные продажи по регионам

```python
data['total_copies_sold'] = data[['na_sales','eu_sales','jp_sales', 'other_sales']].sum(axis = 1)
```

```python
# Посмотрим, что у нас осталось
temp = data.copy() 
list_c = ['name', 'platform', 'year_of_release', 'genre', 'critic_score', 'user_score', 'rating']
print(temp.info())
for col_l in list_c:
  print('-------------------------')
  print(col_l, temp[col_l].sort_values().unique())
  print(col_l,': кол-во NaN',temp[col_l].isna().sum(),
        ', процент NaN', round(temp[col_l].isna().sum()/len(temp)*100, 2),'%')
```

Пропуски могут быть «Game Experience May Change During Online Play» - онлайн рейтинг для игр, содержание которых может модифицироваться.

```python
data['rating'] = data['rating'].fillna("GEMCDOP") 
```

На графике заметны интервалы "жизни" платформы: 9-10 лет. Примем актуальный период - период популярности платформы - равный 5 лет.

#### Проанализируем платформы, пользующиеся популярностью за последние 5 лет. Выберем несколько потенциально прибыльных из них.

Сформируем список платформ, на которых игры начали выпускаться раньше 2011г: 2016 - актуальный период

```python
data_potential_platform_pivot = data.pivot_table(index=['platform', 'year_of_release'],
                                                                   values='total_copies_sold', aggfunc=['sum'])
data_potential_platform_pivot.columns = ['sum_total_copies']
old_platforms = data_potential_platform_pivot[data_potential_platform_pivot.index.get_level_values(1) < 2011].index.get_level_values(0)
old_platforms = list(old_platforms.unique())
```

```python
potential_platforms = data[~data['platform'].isin(old_platforms)]
potential_platforms_pivot = potential_platforms.pivot_table(index=['platform', 'year_of_release'],
                                                            values='total_copies_sold', aggfunc=['sum']).reset_index()
potential_platforms_pivot.columns = ['platform', 'year_of_release', 'total_copies_sold']
plt.figure(figsize=(20,8))
sns.barplot(data=potential_platforms_pivot, x='year_of_release', y='total_copies_sold', hue='platform', palette='dark')
plt.title('Распределение продаж за последние 5 лет самых популярных платформ', fontsize=15)
plt.legend(fontsize='large', shadow=True)
plt.xlabel('Год выпуска', fontsize=15)
plt.ylabel('Копий продано, мнл', fontsize=15)
plt.show()
```

Потенциально прибыльные платформы: PS4, XOne, 3DS

```python
actual_platform = ['PS4','XOne','3DS']
plt.figure(figsize=(20,8))
sns.boxplot(data=potential_platforms.loc[data['platform'].isin(actual_platform)], x='total_copies_sold', y='platform', palette='Set3', orient='h')
plt.title('Диаграмма размаха прибыли от игр для платформ: PS4, XOne, 3DS', fontsize=20)
plt.xlabel('Продано копий, млн', fontsize=15)
plt.ylabel('Платформа', fontsize=15)
plt.show()
```

```python
data_top_platform = potential_platforms.loc[data['platform'].isin(actual_platform)]
data_top_platform_pivot = data_top_platform.pivot_table(index='platform', values='total_copies_sold', aggfunc='sum').reset_index()
data_top_platform_pivot.columns = ['platform', 'sum_total_copies']
plt.figure(figsize=(20,8))
explode = [0, 0.1, 0]
plt.pie(data_top_platform_pivot['sum_total_copies'],
        labels=data_top_platform_pivot['platform'],
        autopct='%1.1f%%',
        radius=1,
        shadow=True,
        explode=explode,
        wedgeprops={'lw':1, 'ls':'--','edgecolor':"k"},
        textprops={'size': '20'})
plt.title('Доли продаж актуальных платформ', fontsize=15)
plt.show()
```

```python
data_ps4_games = data[data['platform'] == 'PS4'].reset_index()
```

```python
def copies_sold(value):
    res = "Топ - более 2млн копий"
    if value <= 2:
        res = "Обычные - до 2млн копий"
    return res
```

### Проверка гипотез

1. Нулевая гипотеза H0: Средние пользовательские рейтинги платформ XboxOne и PC отличаются
2. Альтернативная гипотеза H1: Средние пользовательские рейтинги платформ XboxOne и PC не отличаются

```python
alpha = 0.05
```

```python
data_xone = top_data.query('platform == "XOne"').dropna(subset=['user_score'])['user_score']
data_pc = top_data.query('platform == "PC"').dropna(subset=['user_score'])['user_score']
```

```python
# проверка гипотезы о равенстве средних
results = st.ttest_ind(data_xone, data_pc, equal_var = False)
print(f'pvalue = {results.pvalue}')
if results.pvalue < alpha:
    print('Отвергаем гипотезу')
else:
    print('Не отвергаем гипотезу')
```

1. Нулевая гипотеза H0: Средние пользовательские рейтинги жанров action и sports отличаются
2. Альтернативная гипотеза H1: Средние пользовательские рейтинги жанров action и sports не отличаются

```python
data_action = top_data.query('genre == "Action"').dropna(subset=['user_score'])['user_score']
data_sports = top_data.query('genre == "Sports"').dropna(subset=['user_score'])['user_score']
```

```python
# проверка гипотезы о равенстве средних
results = st.ttest_ind(data_action, data_sports, equal_var = False)
print(f'pvalue = {results.pvalue}')
if results.pvalue < alpha:
    print('Отвергаем гипотезу')
else:
    print('Не отвергаем гипотезу')
```

Вывод:
1. Средние пользовательские рейтинги платформ XboxOne и PC в основном не отличаются
2. Средние пользовательские рейтинги игр жанров Action и Sports в основном отличаются

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
sns.boxplot(data=top_data[(top_data['platform']=='XOne') | (top_data['platform']=='PC')],
            x='platform', y='user_score', showfliers = False, ax=ax1)
sns.boxplot(data=top_data[(top_data['genre']=='Action') | (top_data['genre']=='Sports')],
            x='genre', y='user_score', showfliers = False, ax=ax2)
ax1.set_title('Диаграммы размаха платформ Xone и PC')
ax2.set_title('Диаграммы размаха жанров Action и Sports')
plt.show()
```
